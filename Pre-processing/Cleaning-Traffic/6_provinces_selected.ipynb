Combine the year (3 files)

```python
import pandas as pd
import os

# Path to the folder containing the selected CSV files
input_folder = "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Weekend_selected_province(5)"
output_folder = "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Combined_by_year(6)"

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Initialize dictionaries to store dataframes for each year
yearly_data = {2022: [], 2023: [], 2024: []}

# Iterate over each file in the input folder
for file_name in os.listdir(input_folder):
    if file_name.endswith(".csv"):  # Only process CSV files
        # Construct the full file path
        file_path = os.path.join(input_folder, file_name)
        
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(file_path)
        
        # Extract the year from the 'Date' column (assuming it's in 'YYYY-MM-DD' format)
        df['Year'] = pd.to_datetime(df['Date']).dt.year
        
        # Append data to the corresponding year list
        for year in [2022, 2023, 2024]:
            if year in df['Year'].values:
                yearly_data[year].append(df[df['Year'] == year])

# Combine and save the data for each year into separate files
for year, data_frames in yearly_data.items():
    if data_frames:  # If there is data for the year
        combined_df = pd.concat(data_frames, ignore_index=True)
        output_file_path = os.path.join(output_folder, f"combined_{year}.csv")
        combined_df.to_csv(output_file_path, index=False)
        print(f"Combined data for {year} saved to: {output_file_path}")
```

จากผลลัพธ์ที่ได้ แสดงให้เห็นว่า:

การทำความสะอาดข้อมูล:

	•	วันที่ในไฟล์ combined_2024.csv อยู่ในช่วง 2024-03-02 ถึง 2024-11-30 ซึ่งเป็นช่วงวันที่ที่ถูกต้องตามปี 2024 ที่เราต้องการ.
	2.	สถิติเกี่ยวกับความเร็ว (Speed Statistics):
	•	ความเร็ว (speed_kmh):
	•	ค่าเฉลี่ย (mean) ของความเร็ว = 26.87 km/h
	•	ความเร็วสูงสุด (max) อยู่ที่ 200 km/h ซึ่งแสดงถึงความเร็วที่สูงมาก.
	•	จากการตรวจสอบนี้, ค่าความเร็วสูงกว่า 180 km/h ถือว่าเป็นค่าผิดปกติ, ดังนั้นมีการแจ้งเตือนว่าความเร็วบางค่ามีความสูงเกินไป (ซึ่งอาจบ่งชี้ว่ามีข้อผิดพลาดในการเก็บข้อมูล).
	3.	การตรวจสอบค่าความเร็วสูงเกินไป:
	•	ได้มีการแจ้งเตือนว่า “⚠️ Speed contains values greater than 180 km/h, which may indicate errors.” ซึ่งแสดงว่ามีข้อมูลที่อาจผิดปกติหรือไม่ได้รับการเก็บข้อมูลอย่างถูกต้อง.
	•	การแจ้งเตือนนี้ช่วยให้คุณสามารถตรวจสอบข้อมูลเหล่านี้เพิ่มเติมหากต้องการ.
	4.	บันทึกข้อมูลที่ทำความสะอาดแล้ว:
	•	ข้อมูลที่ทำการตรวจสอบและทำความสะอาดเสร็จแล้วจะถูกบันทึกในไฟล์ใหม่ที่ชื่อว่า combined_2024.csv ซึ่งถูกบันทึกไว้ในโฟลเดอร์ Processed_data_by_year ตามเส้นทางที่ระบุ.

ข้อมูลที่ได้รับการทำความสะอาดแล้ว:
	•	ข้อมูลซ้ำถูกลบออก
	•	ข้อมูลที่หาย (missing data) ถูกแทนที่ด้วยค่าเฉลี่ย
	•	ข้อมูลที่มีความเร็วไม่สมเหตุสมผล (ต่ำกว่า 0 หรือมากกว่า 200 km/h) ถูกกรองออก
	•	ข้อมูลวันที่อยู่ในช่วงที่ถูกต้อง
	•	ข้อมูลจังหวัด Province_en ได้รับการตรวจสอบและแสดงค่าที่ไม่ถูกต้อง

```python
import pandas as pd
import numpy as np
import os

INPUT_FILES = [
    "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Combined_by_year(6)/combined_2022.csv",
    "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Combined_by_year(6)/combined_2023.csv",
    "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Combined_by_year(6)/combined_2024.csv"
]

# Function to check and clean data
def clean_data(df):

    duplicates = df.duplicated(subset=['lat', 'lon', 'Date', 'speed_kmh', 'Province_en']).sum()
    print(f"Number of duplicates: {duplicates}")
    df = df.drop_duplicates(subset=['lat', 'lon', 'Date', 'speed_kmh', 'Province_en'])
    

    missing_data = df.isnull().sum() / len(df) * 100
    print("Missing data percentage:\n", missing_data)
    

    if missing_data.any():
        df['speed_kmh'].fillna(df['speed_kmh'].mean(), inplace=True)
        df['lat'].fillna(df['lat'].mean(), inplace=True)
        df['lon'].fillna(df['lon'].mean(), inplace=True)


    try:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    except Exception as e:
        print(f"⚠️ Error converting Date column: {e}")
        return None
    
    date_min, date_max = df['Date'].min(), df['Date'].max()
    print(f"Date range: {date_min} to {date_max}")
    

    if date_min.year < 2022 or date_max.year > 2024:
        print("⚠️ Data contains dates outside the range of 2022-2024")

    df = df[(df['speed_kmh'] >= 0) & (df['speed_kmh'] <= 200)]
    

    valid_provinces = ['Bangkok', 'Nonthaburi', 'Samut Sakhon', 'Pathum Thani', 'Nakhon Pathom']
    invalid_provinces = df[~df['Province_en'].isin(valid_provinces)]
    if not invalid_provinces.empty:
        print("⚠️ Invalid Province_en values found:", invalid_provinces['Province_en'].unique())


    speed_stats = df['speed_kmh'].describe()
    print(f"Speed statistics:\n{speed_stats}")
    

    if speed_stats['max'] > 180:
        print("⚠️ Speed contains values greater than 180 km/h, which may indicate errors.")

    return df if not df.empty else None


def process_files(files):
    for file_path in files:
        try:
            print(f"Processing {file_path}...")
            df = pd.read_csv(file_path)

            # Clean data by calling the clean_data function
            cleaned_df = clean_data(df)
            
            if cleaned_df is not None:
                # Save the cleaned data into a new file with the same name in the output directory
                output_dir = "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Processed_data_by_year"
                os.makedirs(output_dir, exist_ok=True)
                
                output_path = os.path.join(output_dir, os.path.basename(file_path))
                cleaned_df.to_csv(output_path, index=False, encoding="utf-8-sig")
                print(f"Finished cleaning and saved: {output_path}")
            else:
                print(f"Skipping {file_path} due to empty or invalid data")

        except Exception as e:
            print(f"❌ Error processing {file_path}: {e}")


process_files(INPUT_FILES)
```

Calculate index

```python
import pandas as pd
import os

INPUT_FILES = [
    "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Processed_data_by_year (7)/combined_2022.csv",
    "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Processed_data_by_year (7)/combined_2023.csv",
    "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Processed_data_by_year (7)/combined_2024.csv"
]

# Function to calculate Traffic Jam Index
def calculate_traffic_jam_index(df):
    # Convert Date to datetime format
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

    # Filter out invalid data where Date could not be parsed
    df = df.dropna(subset=['Date'])

    # Group by Date and Province and calculate Traffic Jam Index
    df['Traffic_jam_index'] = df.groupby(['Date', 'Province_en'])['speed_kmh'].transform(
        lambda x: (x == 0).sum() / len(x)  # number of vehicles with speed=0 / total vehicles
    )
    
    return df[['Date', 'speed_kmh', 'Province_en', 'Traffic_jam_index']]

# Function to process files
def process_files(files):
    for file_path in files:
        try:
            print(f"Processing {file_path}...")
            df = pd.read_csv(file_path)

            # Calculate Traffic Jam Index
            cleaned_df = calculate_traffic_jam_index(df)

            # Save the cleaned data into a new file with the same name but in a new directory
            output_dir = "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Processed_traffic_jam_index"
            os.makedirs(output_dir, exist_ok=True)
            
            output_path = os.path.join(output_dir, os.path.basename(file_path))
            cleaned_df.to_csv(output_path, index=False, encoding="utf-8-sig")
            print(f"Finished processing and saved: {output_path}")

        except Exception as e:
            print(f"Error processing {file_path}: {e}")

# Call the process_files function
process_files(INPUT_FILES)
```
