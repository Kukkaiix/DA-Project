```python
import pandas as pd
import os
from geopy.distance import geodesic
from scipy.spatial.distance import cdist
import numpy as np

INPUT_PATH = "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Weekend_3_10_11"
TAMBON_PATH = "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Tambon/Tambon Data.csv"
OUTPUT_PATH = "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Tambon/Weekend_3_10_11_with_province"


try:
    tambon_df = pd.read_csv(TAMBON_PATH)
    print(f"Successfully loaded Tambon data from {TAMBON_PATH}")
except Exception as e:
    print(f" Error loading Tambon data: {str(e)}")
    tambon_df = None

if tambon_df is not None:
    tambon_coords = list(zip(tambon_df['LAT'], tambon_df['LONG']))
    tambon_provinces = tambon_df['CHANGWAT_E'].tolist()
    print(f" Loaded {len(tambon_coords)} coordinates for matching.")
else:
    tambon_coords, tambon_provinces = [], []


def get_nearest_province(lat, lon):
    """find Province_en from lat, lon with min distance"""

    lat_lon_input = np.array([[lat, lon]])
    tambon_coords_np = np.array(tambon_coords)
    distances = cdist(lat_lon_input, tambon_coords_np, metric='euclidean')  # Euclidean distance 

    #finr the min distance
    min_index = np.argmin(distances)
    nearest_province = tambon_provinces[min_index]
    
    return nearest_province


def add_province_to_data(input_file, output_file):
    try:
        print(f"ğŸ”„ Processing file: {input_file}")
        
        df = pd.read_csv(input_file)
        
        
        if "lat" not in df.columns or "lon" not in df.columns:
            print(f"âš ï¸ Missing lat or lon column in {input_file}")
            return

        print(f"âœ… Found 'lat' and 'lon' columns in {input_file}. Proceeding with province matching.")

        
        df['Province_en'] = df.apply(lambda row: get_nearest_province(row['lat'], row['lon']), axis=1)
        print(f"ğŸŒ Added 'Province_en' column to {input_file}")
        
        
        os.makedirs(output_file, exist_ok=True)

        
        output_path = os.path.join(output_file, os.path.basename(input_file))
        df.to_csv(output_path, index=False, encoding="utf-8-sig")
        
        print(f"File saved: {output_path}")
    except Exception as e:
        print(f"Error processing {input_file}: {str(e)}")


def process_all_files(input_dir, output_dir):
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            if file.endswith(".csv"):
                file_path = os.path.join(root, file)
                add_province_to_data(file_path, output_dir)


print(f" Starting file processing from {INPUT_PATH}...")
process_all_files(INPUT_PATH, OUTPUT_PATH)
print(f" Finished processing all files.")
```
