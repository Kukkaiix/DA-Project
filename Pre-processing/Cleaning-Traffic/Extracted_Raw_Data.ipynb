Extracted The Raw Data from .tar.bz2 --> csv.out --> .csv

```python
import os
import tarfile
src_dir = "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Road_Raw_Data"

extract_root = "/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Extracted_bz2"
os.makedirs(extract_root, exist_ok=True)

archives = [f for f in os.listdir(src_dir) if f.endswith(".tar.bz2")]
print(f"Found {len(archives)} archive(s):", archives)

for name in archives:
    file_path = os.path.join(src_dir, name)
    out_dir = os.path.join(extract_root, name.replace(".tar.bz2", ""))
    os.makedirs(out_dir, exist_ok=True)

    print(" Extracting:", file_path, "->", out_dir)
    try:
        with tarfile.open(file_path, "r:bz2") as tar:
            tar.extractall(out_dir)
        print("âœ… Done")
    except Exception as e:
        print("âŒ Error:", e)

    
for dp, dn, fns in os.walk(extract_root):
    for f in fns:
        print("Extracted file:", os.path.join(dp, f))

renamed = 0
for dp, dn, fns in os.walk(extract_root):
    for f in fns:
        if f.endswith(".csv.out"):
            old = os.path.join(dp, f)
            new = os.path.join(dp, f[:-4])  # à¸•à¸±à¸” .out à¸­à¸­à¸
            os.rename(old, new)
            renamed += 1
print(f"âœ… Renamed {renamed} files")
```

```python
import os
import pandas as pd
from pathlib import Path


ROOT = Path("/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Extracted_bz2")

COLS = [
    "vehicle_id","gps_valid","lat","lon","timestamp",
    "speed_kmh","heading_deg","for_hire_light","engine_acc"
]


csv_files = list(ROOT.rglob("*.csv"))
print("Found CSV files:", len(csv_files))

for fp in csv_files:
    try:

        df = pd.read_csv(fp, header=None, names=COLS, usecols=range(9), on_bad_lines="skip")


        df.to_csv(fp, index=False)
        print(f"âœ… Added header to {fp.name}")
    except Exception as e:
        print(f"âš ï¸ {fp.name}: {e}")
```
