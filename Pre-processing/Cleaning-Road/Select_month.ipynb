```python
import pandas as pd
from pathlib import Path

print("start")

# กำหนดที่อยู่ของโฟลเดอร์
ROOT = Path("/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Extracted_bz2")
SELECTED_MONTH_DIR = Path("/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Selected Month/3_10_11_Data")
SELECTED_MONTH_DIR.mkdir(parents=True, exist_ok=True)

# ตั้งชื่อคอลัมน์ตามสเปก iTIC
COLS = [
    "vehicle_id", "gps_valid", "lat", "lon", "timestamp",
    "speed_kmh", "heading_deg", "for_hire_light", "engine_acc"
]

# คอลัมน์ที่อยากเก็บ
KEEP_COLS = ["lat", "lon", "timestamp", "speed_kmh"]

# กำหนดเดือนที่ต้องการ
SELECTED_MONTHS = [3, 10, 11]

# หาไฟล์ .csv ทั้งหมดใน ROOT
csv_files = list(ROOT.rglob("*.csv"))
print(f"Found {len(csv_files)} CSV files.")

# ฟังก์ชันกรองข้อมูลที่เดือนที่ต้องการ
def filter_by_month(df, months):
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    df['month'] = df['timestamp'].dt.month
    return df[df['month'].isin(months)]

# ลูปทุกไฟล์ .csv
for fp in csv_files:
    try:
        # อ่านไฟล์
        df = pd.read_csv(fp, header=None, names=COLS, usecols=range(9), on_bad_lines="skip")

        # กรองข้อมูลตามเดือนที่ต้องการ
        df_filtered = filter_by_month(df, SELECTED_MONTHS)

        # เก็บเฉพาะคอลัมน์ที่ต้องการ
        df_filtered = df_filtered[KEEP_COLS]

        # เปลี่ยนชื่อคอลัมน์
        df_filtered = df_filtered.rename(columns={"lat": "Latitude", "lon": "Longitude"})

        # แยก Date จาก timestamp (ในรูปแบบ YY-MM-DD)
        df_filtered['Date'] = df_filtered['timestamp'].dt.strftime('%y-%m-%d')

        # เก็บเฉพาะคอลัมน์ที่ต้องการ
        df_filtered = df_filtered[["Latitude", "Longitude", "Date", "speed_kmh"]]

        # สร้างชื่อไฟล์ใหม่ และบันทึกไฟล์ที่กรองแล้ว
        output_file = SELECTED_MONTH_DIR / fp.name
        df_filtered.to_csv(output_file, index=False)
        print(f"✅ Saved filtered data to {output_file}")

    except Exception as e:
        print(f"⚠️ Error processing {fp.name}: {e}")
    
print("end the process")
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from pathlib import Path
from tqdm import tqdm  # Import tqdm for the progress bar

print("Start processing...")

# กำหนดที่อยู่ของโฟลเดอร์
SELECTED_MONTH_DIR = Path("/Users/kukkaii/Documents/DA PRoject/Clean_RoadData/Selected Month/3_10_11_Data")

# หาไฟล์ .csv ทั้งหมดใน SELECTED_MONTH_DIR
csv_files = list(SELECTED_MONTH_DIR.rglob("*.csv"))
print(f"Found {len(csv_files)} CSV files.")

# สร้างตัวแปร geolocator สำหรับการใช้ Nominatim API
geolocator = Nominatim(user_agent="geoapiExercises")

# ฟังก์ชันที่ใช้ Nominatim API เพื่อแปลง Lat/Lon เป็นชื่อจังหวัด
def get_province_en(lat, lon):
    try:
        location = geolocator.reverse((lat, lon), language='en', exactly_one=True)
        address = location.raw.get('address', {})
        province = address.get('state', '')  # เอาชื่อจังหวัดจากผลลัพธ์
        return province
    except Exception as e:
        print(f"Error in geocoding {lat}, {lon}: {e}")
        return None

# ลูปทุกไฟล์ .csv
for fp in tqdm(csv_files, desc="Processing files", unit="file"):
    try:
        # อ่านไฟล์
        df = pd.read_csv(fp)

        # เพิ่มคอลัมน์ 'Province_en' โดยใช้ข้อมูล Lat/Lon
        df['Province_en'] = df.apply(lambda row: get_province_en(row['Latitude'], row['Longitude']), axis=1)

        # บันทึกไฟล์ที่กรองแล้ว
        df.to_csv(fp, index=False)
        print(f"✅ Processed and saved data to {fp.name}")

    except Exception as e:
        print(f"⚠️ Error processing {fp.name}: {e}")

print("End processing")
```
